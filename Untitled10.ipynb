```python
# Install required libraries
!pip install flask pyngrok scikit-learn pandas

# Import required modules
from flask import Flask, request, jsonify
from pyngrok import ngrok
import threading
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder

# ---------------------
# Step 1: Prepare data and train model
data = pd.DataFrame({
    'location': ['nungwi', 'kendwa', 'pemba', 'kizimkazi'],
    'sst': [27.3, 26.9, 27.8, 26.4],
    'chlorophyll': [1.0, 0.8, 1.2, 0.5],
    'fish_abundance': ['High', 'Medium', 'High', 'Low']
})

le = LabelEncoder()
data['fish_abundance_encoded'] = le.fit_transform(data['fish_abundance'])

X = data[['sst', 'chlorophyll']]
y = data['fish_abundance_encoded']

model = GradientBoostingClassifier()
model.fit(X, y)

# ---------------------
# Step 2: Define chatbot function
def chatbot_with_model(user_input):
    user_input = user_input.lower()
    for i in range(len(data)):
        location = data.loc[i, 'location']
        if location in user_input:
            sst = data.loc[i, 'sst']
            chlorophyll = data.loc[i, 'chlorophyll']
            pred = model.predict([[sst, chlorophyll]])
            label = le.inverse_transform(pred)[0]
            return f"Eneo la {location.capitalize()} lina samaki {label.lower()} leo (SST: {sst}Â°C, Chlorophyll: {chlorophyll})."
    if "samaki" in user_input and ("wapi" in user_input or "sehemu" in user_input or "eneo" in user_input):
        high_areas = data[data['fish_abundance'] == 'High']['location'].str.capitalize().tolist()
        return f"Leo samaki wengi wapo katika maeneo haya: {', '.join(high_areas)}."
    return "Samahani, sijakuelewa. Tafadhali uliza kuhusu eneo au samaki."

# ---------------------
# Step 3: Setup Flask and pyngrok
app = Flask(__name__)

@app.route('/chat', methods=['POST'])
def chat():
    user_input = request.json.get('message')
    if not user_input:
        return jsonify({'response': 'Hakuna ujumbe umetumwa.'}), 400
    response = chatbot_with_model(user_input)
    return jsonify({'response': response})

def run_app():
    app.run()

# Set your ngrok authtoken here!
ngrok.set_auth_token("30dN35f6xYKikbFYpd5jOABSOd6_6mo4B69kFhrknm33EU2yv")

# Start Flask in background thread
thread = threading.Thread(target=run_app)
thread.start()

# Open ngrok tunnel to port 5000
public_url = ngrok.connect(5000)
print(f"ðŸŒ Public URL: {public_url}")
```
